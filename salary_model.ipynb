```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score

# 1. Load data
df = pd.read_csv("salary_data.csv")

X = df[['exp']]
y = df['salary']

# 2. Train model
model = LinearRegression()
model.fit(X, y)

# 3. Predict on training data
y_pred = model.predict(X)

# 4. Evaluate
rmse = root_mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R2 Score: {r2:.2f}")

# 5. Plot
plt.scatter(X, y, marker='x', color='red', label='Real Data')
plt.plot(X, y_pred, color='black', label='Predicted')
plt.legend()
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary Prediction (Linear Regression)')
plt.grid()
plt.show()

# 6. Predict for new input
exp = int(input("Enter years of experience: "))
eexp=pd.DataFrame([[exp]],columns=['exp'])
pred_salary = model.predict(eexp)[0]
print(f"Predicted salary for {exp} years of experience: {pred_salary:.2f}")
```
